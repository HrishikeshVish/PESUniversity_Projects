```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("input.csv", low_memory=False);
```

```python
print("Total rows: {0}".format(len(data)))
```

```python
cols = list(data);
print(len(cols))
```

```python
valcounts = {};
totcol = len(data);
for i in cols:
    valcounts[i] = totcol-data[i].count()

for key in valcounts:
    print(key+" "+str(valcounts[key])+"\n")
```

```python
perdic = {}
for key in valcounts:
    per = totcol - valcounts[key]
    per = per/totcol
    perdic[key] = 1-per;
for key in perdic:
    print(key+" "+str(perdic[key])+"\n")
```

```python
for key in perdic:
    if(perdic[key]>=0.5):
        data = data.drop(key, 1)
print(len(list(data)))
```

```python
cols = list(data)
for i in cols:
    print(i + "\n")
```

```python
data.to_csv('input.csv');
```
