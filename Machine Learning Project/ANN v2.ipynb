```python
import pandas as pd
import numpy as np
import math
import random
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def sigmoid_der(x):
    return sigmoid(x)*(1-sigmoid(x))

def softmax(A):
    expA = np.exp(A)
    return expA / expA.sum(axis=1, keepdims=True)
```

```python
data = pd.read_csv("ANN.csv", low_memory = False).drop('Unnamed: 0', 1);
cols = list(data);
print(cols, len(cols));

labels = np.array(data['P. Habitable Class']).reshape(len(data), 1)
print(labels)
feature_set = np.array(data.drop('P. Habitable Class', 1))
print(feature_set)
print(feature_set.shape)
```

```python
np.random.seed(42)
np.set_printoptions(precision=3)

attributes = feature_set.shape[1]
hidden_nodes = 22

wh = np.random.rand(attributes,hidden_nodes)
# bh = np.random.randn(hidden_nodes)

wo = np.random.rand(hidden_nodes, 1)
# bo = np.random.rand(1)
lr = 0.05

for epoch in range(20):
    # feedforward
    zh = np.dot(feature_set, wh)# + bh
    ah = sigmoid(zh)

    zo = np.dot(ah, wo)# + bo
    ao = sigmoid(zo)

    # Phase1 =======================

    error_out = ((1 / 2) * (np.power((ao - labels), 2)))
#     print(error_out.sum())

    dcost_dao = ao - labels
    dao_dzo = sigmoid_der(zo)
    dzo_dwo = ah

    dcost_wo = np.dot(dzo_dwo.T, dcost_dao * dao_dzo)
    
#     dcost_bo = dcost_dzo

    # Phase 2 =======================

    # dcost_w1 = dcost_dah * dah_dzh * dzh_dw1
    # dcost_dah = dcost_dzo * dzo_dah
    dcost_dzo = dcost_dao * dao_dzo
    dzo_dah = wo
    dcost_dah = np.dot(dcost_dzo , dzo_dah.T)
    dah_dzh = sigmoid_der(zh)
    dzh_dwh = feature_set
    dcost_wh = np.dot(dzh_dwh.T, dah_dzh * dcost_dah)
    
#     dcost_bh = dcost_dah * dah_dzh

    # Update Weights ================

    wh -= lr * dcost_wh
#     bh -= lr * dcost_bh.sum()
    
    wo -= lr * dcost_wo
#     bo -= lr * dcost_bo.sum()
    
    print("Epoch:\t{}\tError:\t{}".format(epoch, error_out.sum()))
#     print(wh)
#     print(wo)
print("\nLossy Accuracy: {}%".format((feature_set.shape[0] - error_out.sum()) * 100 / feature_set.shape[0]))
predicted = error_out.tolist()
predicted = [1 if i[0] > 0.1 else 0 for i in predicted]
actual = [i[0] for i in labels]
vals = {"tp": 0, "tn": 0, "fp": 0, "fn": 0}
n = feature_set.shape[0]
for i in range(n):
    if predicted[i]:
        if actual[i]:
            vals["tp"] += 1
        else:
            vals["fp"] += 1
    else:
        if actual[i]:
            vals["fn"] += 1
        else:
            vals["tn"] += 1

TP, TN, FP, FN = vals.values()
accuracy = (TP+TN)/(TP+FP+FN+TN)
precision = TP/(TP+FP)
recall = TP/(TP+FN)
fscore = 2*(recall*precision)/(recall+precision)

print("\n\nAccuracy: {}, Precision: {}, Recall: {}, F1 Score: {}".format(accuracy, precision, recall, fscore))
```
