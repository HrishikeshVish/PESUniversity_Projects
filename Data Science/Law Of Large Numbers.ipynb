```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import random

X=[1,2,3,4,5,6]
trials= [i for i in range(10,2000,50)]
values=[]
means=[]
for i in trials:
    values=[]
    for j in range(i):
        values.append(random.sample(X,1))
    means.append(np.mean(values))
    
plt.plot(trials, means)
plt.hlines(xmin=0,xmax=2000,y=3.5)
plt.ylim(1,6)
plt.show()

#As seen in the output the value of mean converges to expectation
#Hence if u roll a die a large number of times in this case 2000 times the mean of the numbers that show up, their mean will 
#be close to the expectation of the Random Variable X(in this case 1 to 6)
```
