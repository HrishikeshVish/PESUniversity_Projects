```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pandas.tools.plotting import scatter_matrix

df = pd.read_csv('height-weight.csv') #creates a data frame from the given file
print(df.describe()) #GIVES THE FUVE NUMBER SUMMARY OF THE DATA SET
print(df.dtypes) #specifies the data type of each column
print(df.head()) #diaplays first five rows
print(df.tail()) #displays last five rows
```

```python
pd.scatter_matrix(df)
plt.show()
```

```python
df.height.hist()  # OR plt.hist(df.height)
plt.xlabel('Height')
plt.ylabel('Freq')
plt.show()
```

```python
freq, bins, patches = plt.hist(df.height)
print( freq,bins) #by default python divides it into 10 equal bins
```

```python
plt.figure(figsize=[20,20]) #change figure size
bp = plt.boxplot(df.height, showmeans=True)
outliers = bp['fliers'][0].get_ydata()
print(outliers)
plt.show()
```
