```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from math import sqrt

df = pd.read_csv('height-weight.csv')

print('Pop_mean',np.mean(df.height) )
print('Pop SD', np.std(df.height))

def sampling(sample_size, no_of_samples):
    print('------------------------------')
    means = [] #means of all 100 samples
    for i in range(no_of_samples):
        x = df.height.sample(sample_size)
        means.append(np.mean(x))
    print('Mean of means ', np.mean(means))
    print('SD with Sample_Size',sample_size,'is:', np.std(means, ddof=1))
    plt.hist(means)
        

sampling(50,100)
sampling(100,100)
sampling(1000,100)
print('expected SD', np.std(df.height)/sqrt(100))
```
